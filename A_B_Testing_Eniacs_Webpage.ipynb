# A/B Test Setup



 *  **Test Duration:** 14 days
 *  **Visitors per Version:** ~25,000
 *  **Homepage Versions Tested:** 4













# Key Metrics


*   **Primary Metric:** Click-Through Rate (CTR) for the main call-to-action button.
*   **Secondary Metrics:** Drop-Off Rate (users leaving homepage) and Homepage Return Rate (users revisiting).



# Formulating the Hypotheses



*   **Null Hypothesis (H0):** Changes in button design do not significantly influence CTR or sales.
*   **Alternative Hypothesis (H1):** Changes in button design significantly influence CTR or sales.

# Calculate test Result

## Chi-square-Test

Ater conducting the chi-square-test, we could prove that our pvalue is significantly smaler than our alpha which leads us to reject our null Hypothesis.

```python
# Calculate Chi-Square-Test Stats

chisq, pvalue, df, expected = chi2_contingency(overall_ctr)
```

```python
pvalue

print(f"{pvalue:.100f}")

# pvalue significantly smaller than alpha (0.05).
# We reject the null Hypothesis because we can show that there is a statistical significance in version performance.
```

```python
# Define ciritcal value
critical_value = chi2.ppf(1 - alpha, df)
```

```python
# Interpretation of results

if pvalue < alpha :
  print('The p_value is smaller than alpha')
  print('Siginificant change in Version. We reject H0, because can say with 95% certainty that there is CTR difference in Versions')

else :
  print('The p_value is bigger than alpha')
  print('There is no significant change in Verions. We fail to reject H0, because we can say with 95% certainty that there is no CTR difference in Verisons')
```

### Statistical Significance due to Version B and D performing significantly worse than A and C

```python
# Stats from Chi-square Test

summary = pd.DataFrame({
    'Metric': [
        'Chi-square statistic',
        'p-value',
        'Degrees of freedom',
        'Alpha',
        'Critical value',
        'Decision'
    ],
    'Value': [
        round(chisq, 2),
        f"{pvalue:.3e}",
        df,
        0.05,
        round(critical_value, 2),
        'Reject H0'
    ]
})

summary
```

```python
# @title
# values from your test
chi_stat = chisq        # your test statistic
dof = df                # degrees of freedom
alpha = 0.05            # significance level

# critical value
critical_value = chi2.ppf(1 - alpha, dof)

# x range (IMPORTANT: keep it reasonable for visualization)
x = np.linspace(0, 20, 1000)
y = chi2.pdf(x, dof)

# plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'Chi-square Distribution (df={dof})')

# rejection region
x_fill = np.linspace(critical_value, x.max(), 300)
y_fill = chi2.pdf(x_fill, dof)
plt.fill_between(x_fill, y_fill, alpha=0.3, label=f'Rejection Region (α = {alpha})')

# critical value line
plt.axvline(critical_value, linestyle='--', linewidth=2,
            label=f'Critical Value = {critical_value:.2f}')

# annotate test statistic (since it's far outside the plot)
plt.text(
    critical_value + 1,
    max(y) * 0.7,
    f'Test Statistic = {chi_stat:.2f}\n(out of range)',
    fontsize=11,
    bbox=dict(facecolor='white', alpha=0.8)
)

# styling
plt.title('Chi-Square Test Visualization', fontsize=16)
plt.xlabel('Chi-Square Value', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.legend()
plt.grid(alpha=0.3)

plt.show()
```

# Post-Hoc-test

Now we cannot prove which version is statistically significant in comaprison to Version A. We just know that there IS a significance


Which leads us to conducts a post-hoc-test which compares the different Versions pairiwse to each other

```python
# Post-hoc-Test
# Create Function to compare Versions Pairwise

def pairwise_chi2(df, v1, v2):
    subset = df.loc[[v1, v2], ['No. clicks', 'Not Clicked']]
    chi2, p, _, _ = chi2_contingency(subset)
    return p
```

```python
# For loop to compare all versions pairwise

results = []

version = overall_ctr.index.tolist()

for i in range(len(version)):
    for j in range(i+1, len(version)):
        v1 = version[i]
        v2 = version[j]
        p = pairwise_chi2(overall_ctr, v1, v2)
        results.append((v1, v2, p))
results
```

## Bonferroni Adjustment

Use the Bonferroni adjustment to correct for multiple comparisons and reduce the risk of false positives (Type I errors).


Since testing many pairs increases the chance of finding a “significant” result by luck, the adjustment makes the significance threshold stricter.

```python
# Bonferroni Adjustment

alpha_adjusted = alpha / 6
```

```python
# Convert Data into DataFrame and compare values

comparison = pd.DataFrame(results, columns=['Version 1', 'Version 2', 'p-value'])

comparison['Significant'] = comparison['p-value'] < alpha_adjusted

# format correctly
comparison['p-value'] = comparison['p-value'].apply(lambda x: f"{x:.3e}")

comparison
```
